```
# Código de simulación para el punto 2.3
import numpy as np
np.set_printoptions(precision=6, suppress=True)

def system1(x):
    N = len(x)
    y = np.zeros(N)
    for n in range(N):
        xn = x[n]
        xn1 = x[n-1] if n-1 >= 0 else 0.0
        yn1 = y[n-1] if n-1 >= 0 else 0.0
        y[n] = xn/3.0 + 2.0*xn1 - yn1
    return y

def system2(x):
    return np.cumsum(x**2)

def system3(x):
    padded = np.pad(x, (1,1), mode='constant', constant_values=0.0)
    y = np.zeros_like(x)
    for n in range(len(x)):
        window = padded[n:n+3]
        y[n] = np.median(window)
    return y

def system4_factory(A=2.0, B=1.0):
    def H(x):
        return A * x + B
    return H

def test_linearity(H, x1, x2, a=2.0, b=-1.5):
    left = H(a*x1 + b*x2)
    right = a*H(x1) + b*H(x2)
    diff = left - right
    return np.allclose(left, right, atol=1e-8), diff

def test_time_invariance(H, x, k=3):
    N = len(x)
    x_shift = np.concatenate((np.zeros(k), x))[:N]
    y_original = H(x)
    y_shifted = H(x_shift)
    y_expected = np.concatenate((np.zeros(k), y_original))[:N]
    same = np.allclose(y_shifted, y_expected, atol=1e-8)
    diff = y_shifted - y_expected
    return same, diff

# Señales de prueba
N = 20
t = np.arange(N)
x1 = np.sin(0.2*np.pi*t)
x2 = 0.5 * np.sign(np.sin(0.1*np.pi*t))
a = 2.0; b = -1.5

# Ejecutar pruebas
print("=== Sistema 1 ===")
lin1, diff1 = test_linearity(system1, x1, x2, a=a, b=b)
ti1, td1 = test_time_invariance(system1, x1, k=2)
print("Lineal?", lin1); print("Invariante?", ti1); print("Diff linearity:", diff1); print("Diff TI:", td1)

print("=== Sistema 2 ===")
lin2, diff2 = test_linearity(system2, x1, x2, a=a, b=b)
ti2, td2 = test_time_invariance(system2, x1, k=2)
print("Lineal?", lin2); print("Invariante?", ti2); print("Diff linearity:", diff2); print("Diff TI:", td2)

print("=== Sistema 3 ===")
lin3, diff3 = test_linearity(system3, x1, x2, a=a, b=b)
ti3, td3 = test_time_invariance(system3, x1, k=2)
print("Lineal?", lin3); print("Invariante?", ti3); print("Diff linearity:", diff3); print("Diff TI:", td3)

print("=== Sistema 4 (A=2,B=1) ===")
H4 = system4_factory(A=2.0, B=1.0)
lin4, diff4 = test_linearity(H4, x1, x2, a=a, b=b)
ti4, td4 = test_time_invariance(H4, x1, k=2)
print("Lineal?", lin4, "(lineal solo si B==0)"); print("Invariante?", ti4); print("Diff linearity:", diff4); print("Diff TI:", td4)

```
